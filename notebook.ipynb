```python
import pandas as pd
from pandas import DataFrame
```

```python
def read_jsons():
    vendas = pd.read_json('data/broken_database_1.json')
    marcas = pd.read_json('data/broken_database_2.json')

    return vendas, marcas
```

```python
vendas, marcas = read_jsons()
```

```python
def fix_names(df: DataFrame, coluna: str) -> DataFrame:
    df[coluna] = df[coluna].str.replace('æ', 'a')
    df[coluna] = df[coluna].str.replace('ø', 'o')

    return df
```

```python
vendas = fix_names(vendas, 'nome')
marcas = fix_names(marcas, 'marca')
```

```python
def export_jsons(*args: DataFrame):
    for i, df in enumerate(args):
        df.to_json(f'fixed/fixed_database_{i+1}.json')
```

```python
export_jsons(vendas, marcas)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
df = pd.read_csv('SQLite.9.csv')

# Gráfico de dispersão para vendas x altura
plt.figure(figsize=(10, 6))
plt.scatter(df['vendas'], df['altura'])
plt.title('Vendas x Altura')
plt.xlabel('Vendas')
plt.ylabel('Altura')
plt.grid(True)
plt.show()

# Gráfico de dispersão para vendas x largura
plt.figure(figsize=(10, 6))
plt.scatter(df['vendas'], df['largura'])
plt.title('Vendas x Largura')
plt.xlabel('Vendas')
plt.ylabel('Largura')
plt.grid(True)
plt.show()

# Gráfico de dispersão para vendas x comprimento
plt.figure(figsize=(10, 6))
plt.scatter(df['vendas'], df['comprimento'])
plt.title('Vendas x Comprimento')
plt.xlabel('Vendas')
plt.ylabel('Comprimento')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import numpy as np

# Selecionar as colunas relevantes
cols = ['vendas', 'altura', 'largura', 'comprimento']
df_selected = df[cols]

# Calcular a matriz de correlação
corr_matrix = df_selected.corr()

# Criar o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, vmin=-1, vmax=1)
plt.title('Heatmap de Correlação')
plt.show()
```
